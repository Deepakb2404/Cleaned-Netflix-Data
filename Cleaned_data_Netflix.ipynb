# Task
Analyze the Netflix dataset "netflix_titles.csv" to understand content trends by country, rating, genre, release year, director, and cast.

## Analyze content by country

### Subtask:
Determine the number of movies and TV shows produced by each country and visualize the top countries.


**Reasoning**:
Group the data by country and type and count the occurrences, then display the result.



```python
country_counts = data.groupby(['country', 'type']).size().unstack(fill_value=0)
print(country_counts)
```

**Reasoning**:
Create a pie plot of the top 10 countries by content count, labeling the axes and adding a title.



```python
import matplotlib.pyplot as plt

top_10_countries = country_counts.sum(axis=1).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='pie')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.title('Top 10 Countries by Content Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analyze content by rating

### Subtask:
Explore the distribution of ratings across movies and TV shows and visualize the results.

**Reasoning**:
Group the data by rating and type and count the occurrences, then display the result.

```python
rating_type_distribution = data.groupby(['rating', 'type']).size().unstack(fill_value=0)
print(rating_type_distribution)
```

**Reasoning**:
Visualize the distribution of ratings across movies and TV shows using a stacked bar chart, labeling the axes and adding a title.

```python
rating_type_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.title('Distribution of Ratings Across Movies and TV Shows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analyze content by genre

### Subtask:
Examine the distribution of content across different genres and identify the most popular genres.

**Reasoning**:
Extract all unique genres from the 'listed_in' column, count the occurrences of each genre for both movies and TV shows, and display the result.

```python
genres = data['listed_in'].str.split(', ').explode()
genre_counts = genres.value_counts()
print(genre_counts)
```

## Analyze release year trends

### Subtask:
Investigate the trend of content releases over the years, focusing on the number of movies and TV shows released each year.

**Reasoning**:
Group the data by release year and type and count the occurrences, then display the result.

```python
release_year_counts = data.groupby(['release_year', 'type']).size().unstack(fill_value=0)
print(release_year_counts)
```

## Analyze director productivity

### Subtask:
Identify the directors who have contributed the most content to Netflix.

**Reasoning**:
Count the occurrences of each director in the dataset and display the top 10 directors.

```python
director_counts = data['director'].value_counts()
print("Top 10 Directors by Content Count:")
print(director_counts.head(10))
```

## Analyze cast popularity

### Subtask:
Discover the actors who have appeared in the most movies and TV shows on Netflix.

**Reasoning**:
Extract all unique actors from the 'cast' column, count the occurrences of each actor, and display the top 10 actors.

```python
cast = data['cast'].str.split(', ').explode()
cast_counts = cast.value_counts()
print("Top 10 Actors by Content Count:")
print(cast_counts.head(10))
```

## Summary of Findings

Based on our analysis:

*   **Content by Country:** The United States and India produce the highest number of titles on Netflix.
*   **Content by Rating:** TV-MA and R are the most common ratings for movies, while TV-MA is the most common rating for TV shows.
*   **Content by Genre:** "International Movies" and "Dramas" are the most frequent genres.
*   **Release Year Trends:** The number of releases has generally increased over the years, with a peak around 2018-2019.
*   **Director Productivity:** Directors like Raúl Campos, Jan Suter, and Marcus Raboy have contributed a significant number of titles.
*   **Cast Popularity:** Actors such as Anupam Kher and Shah Rukh Khan have the highest number of appearances in Netflix content.

This analysis provides insights into the content landscape of Netflix based on the provided dataset.
